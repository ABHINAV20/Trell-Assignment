```python
def get_features(correlation_threshold):
    abs_corrs = corr.abs()
    high_correlations = []
    for i in abs_corrs.iteritems():
        if(i[1] > correlation_threshold):
            high_correlations.append(i[0])
    return high_correlations

features = get_features(0.05) 
print(features) 
x = data[features] 
y = data['quality']
```

Except pH and density all features are skewed and hence logarithmic transformation has to be performed.

Since citric acid has lot  of values as 0 so we will take log 1+x for it.

```python
target_name = 'quality'
y = data['quality']
XX = data.drop(labels=[target_name], axis=1)
XX
```

```python
log_feature = ['fixed acidity','volatile acidity','citric acid','residual sugar','chlorides','free sulfur dioxide','total sulfur dioxide','sulphates','alcohol']

for feat in log_feature:
    if feat == 'citric acid':
        XX[feat] += 1
    XX[feat] = np.log(XX[feat])
XX = XX.drop(['residual sugar'], axis=1)
```

# Outlier Detection

```python
for feature in XX.keys():

    # TODO: Calculate Q1 (25th percentile of the data) for the given feature
    Q1 = np.percentile(XX[feature], q=25)
 
    # TODO: Calculate Q3 (75th percentile of the data) for the given feature
    Q3 = np.percentile(XX[feature], q=75)
 
    # TODO: Use the interquartile range to calculate an outlier step (1.5 times the interquartile range)
    interquartile_range = Q3 - Q1
    step = 2.5 * interquartile_range
 
    # Display the outliers
    print("Data points considered outliers for the feature '{}':".format(feature))
    display(data[~((data[feature] >= Q1 - step) & (XX[feature] <= Q3 + step))])
 
outliers = []
# Remove the outliers, if any were specified
good_data = XX.drop(XX.index[outliers]).reset_index(drop = True)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=3)
regressor = LinearRegression()
regressor.fit(x_train,y_train)
```

```python
test_pred = regressor.predict(x_test)
print(test_pred)
```

```python
from sklearn.metrics import mean_squared_error
test_rmse = mean_squared_error(test_pred, y_test) ** 0.5
print(test_rmse)
```

```python
predicted_data = np.round(test_pred)
print(predicted_data)
```

```python
print("Root Mean Squared Error:")
np.sqrt(metrics.mean_squared_error(y_test, test_pred))
```
